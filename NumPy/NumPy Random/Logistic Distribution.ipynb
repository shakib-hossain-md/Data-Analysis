```python
# Ex: Draw 2x3 samples from a logistic distribution with mean 1 and stddev 2.0
from numpy import random
x = random.logistic(loc=1,scale=2,size=(2,3))
print(x)
```

```python
# Visualization of Logistic Distribution
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.kdeplot(random.logistic(size=1000),fill=False)
plt.show()
```

```python
# Difference between Logistic and Normal Distribution
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.kdeplot(random.normal(scale=2, size=1000),fill=False,label='normal')
sns.kdeplot(random.logistic(size=10000),fill=False,label='logistic')
plt.show()
```
