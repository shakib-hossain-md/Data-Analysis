```python
# Visualization of Poisson Distribution
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.displot(random.poisson(lam=2,size=1000),kde=False)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#Generate normal and poisson distributions
data_normal = np.random.normal(loc=50, scale=7, size=1000)
data_poisson = np.random.poisson(lam=50, size=1000)

# Plot using kdeplot with hist=False
sns.kdeplot(data_normal, label='normal', fill=False)
sns.kdeplot(data_poisson, label='poisson', fill=False)

# Display the plot
plt.legend()
plt.show()
```
