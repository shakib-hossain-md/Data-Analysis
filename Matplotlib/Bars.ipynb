```python
# Creating Bars
# With pyplot, you can use the bar() function to draw bar graphs
# Ex: Draw 4 bars.
import matplotlib.pyplot as plt
import numpy as np

x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])

plt.bar(x,y)
plt.show()
```

```python
# The bar() function takes argument that describes the layout of the bars.
import matplotlib.pyplot as plt
import numpy as np

x=["APPLES","BANANS"]
y=[400,350]

plt.bar(x,y)
plt.show()
```

```python
# Horizontal Bars
"""If you want the bars to be displayed horizontally instead of vertically,use the barh() function."""
# Ex: Draw 4 horizontal bars.
import matplotlib.pyplot as plt
import numpy as np

x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])

plt.barh(x,y)
plt.show()
```

```python
# Bar Color
"""The bar() and barh() take the keyword argument color to set the color of the bars."""
# Ex: Draw 4 red bars
import matplotlib.pyplot as plt
import numpy as np

x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])

plt.bar(x,y,color='red')
plt.show()
```

```python
# Color names
"""You can use of 140 supported color name"""
# Ex: Draw 4 hot pink bar
import matplotlib.pyplot as plt
import numpy as np

x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])

plt.bar(x,y,color='hotpink')
plt.show()
```

```python
# Color Hex
# You can use hexadecimal color value.
# Ex:Draw 4 bars with beautiful green color
import matplotlib.pyplot as plt
import numpy as np

x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])

plt.bar(x,y,color='#4CAF50')
plt.show()
```

```python
# Bar width
"""The bar() takes the argument width to set the width of the bars."""
# Ex: Draw 4 very thin bars
import matplotlib.pyplot as plt
import numpy as np

x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])

plt.bar(x,y,width=0.1)
plt.show()

# The default value of the width 0.8
```

```python
# Bar Height
# Note: For horizontal bars, use the height instead of width.
# Draw 4 very thin bars
import matplotlib.pyplot as plt
import numpy as np

x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])

plt.barh(x,y,height=0.1)
plt.show()
```
