```python
# Creating Pie Charts:
# With pyplot,you can use the pie() function to draw pie charts.
# Ex: A simple pie chart.
import matplotlib.pyplot as plt
import numpy as np

y=np.array([35,25,25,15])
plt.pie(y)
plt.show()
```

```python
# Labels
"""Add labels to the pie chart with the labels parameter.
The labels parameter must be an array with one label for each wedge."""
import matplotlib.pyplot as plt
import numpy as np

y=np.array([35,25,25,15])
mylabels=["Apples","Bananas","Cherries","Dates"]
plt.pie(y,labels=mylabels)
plt.show()
```

```python
# Start Angle
"""The default start angle is x-axis,but you can change the start angle by specifing a start angle parameter.
The startangle parameter is defined with an angle in degrees,default degree is 0."""
# Ex: Start the first wedge at 90 degrees.
import matplotlib.pyplot as plt
import numpy as np
y=np.array([35,25,25,15])
mylabels=["Apples","Bananas","Cherries","Dates"]
plt.pie(y,labels=mylabels,startangle=90)
plt.show()
```

```python
# Explode
"""The explode parameter, if specified, and not None, must be an array with one value for each wedge.
Each value represents how far from the center each wedge is displayed:"""
# Ex: Pull the "Apples" wedge 0.2 from the center of the pie.
import matplotlib.pyplot as plt
import numpy as np
y=np.array([35,25,25,15])
mylabels=["Apples","Bananas","Cherries","Dates"]
myexplode=[0.2,0,0,0]
plt.pie(y,labels=mylabels,startangle=90,explode=myexplode)
plt.show()
```

```python
# Shadow
# Add a shadow to the pie chart by setting the shadows parameter to True
# Ex: Add a shadow
import matplotlib.pyplot as plt
import numpy as np
y=np.array([35,25,25,15])
mylabels=["Apples","Bananas","Cherries","Dates"]
myexplode=[0.2,0,0,0]
plt.pie(y,labels=mylabels,startangle=90,explode=myexplode,shadow=True)
plt.show()
```

```python
# Colors
"""You can set the color of each wedge with the colors parameter.
The colors parameter, if specified, must be an array with one value for each wedge:"""
# Ex: Specify a new color for each wedge
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["black", "hotpink", "b", "#4CAF50"]

plt.pie(y, labels = mylabels, colors = mycolors)
plt.show() 
```

```python
# Legend
# To add a list of explanation for each wedge, use the legend() function:
# Ex: Add a legend
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend()
plt.show() 
```

```python
# Legend with header
# To add a header to the legend, add the title parameter to the legend function.
# Ex: Add a legend with a header:
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend(title = "Four Fruits:")
plt.show() 
```
