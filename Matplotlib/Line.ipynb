```python
# Line style
# Ex: Use a dotted line
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3,8,1,10])
plt.plot(ypoints,ls='dotted')
plt.show()
```

```python
# Ex: Use a dashed line
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3,8,1,10])
plt.plot(ypoints,ls='dashed')
plt.show()
```

```python
# Ex: Use shorter syntax
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3,8,1,10])
plt.plot(ypoints,ls=':')
plt.show()
```

```python
# Line Color
# Ex: Set the line color to red.
import matplotlib.pyplot as plt
import numpy as np

ypoints=np.array([3,8,1,10])
plt.plot(ypoints,c='r')
plt.show()
```

```python
# You can also use Hexadecimal color value
# Ex: Plot with a beautiful green line:
import matplotlib.pyplot as plt
import numpy as np

ypoints=np.array([3,8,1,10])
plt.plot(ypoints,c='#4CAF50')
plt.show()
```

```python
# Or any of the 140 supported color names
# Ex: Plot with the color named "hotpink":
import matplotlib.pyplot as plt
import numpy as np

ypoints=np.array([3,8,1,10])
plt.plot(ypoints,c='hotpink')
plt.show()
```

```python
# Line Width
# Ex: Plot with a 20.5pt wide line.
import matplotlib.pyplot as plt
import numpy as np
ypoints=np.array([3,8,1,10])
plt.plot(ypoints,linewidth='20.5')
plt.show()
```

```python
# Multiple Lines
# Ex: Draw two lines by specifing plt.plot() function for each line
import matplotlib.pyplot as plt
import numpy as np

y1=np.array([3,8,1,10])
y2=np.array([6,2,7,11])

plt.plot(y1)
plt.plot(y2)

plt.show()
```

```python
# Ex: Draw two lines by specifing the x-point and y-point value for both lines
import matplotlib.pyplot as plt
import numpy as np

x1=np.array([0,1,2,3])
y1=np.array([3,8,1,10])
x2=np.array([0,1,2,3])
y2=np.array([6,2,7,11])

plt.plot(x1,y1,x2,y2)
plt.show()
```
