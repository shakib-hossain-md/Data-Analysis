```python
# Mark each point with a circle 
import matplotlib.pyplot as plt
import numpy as np
ypoints = np.array([3,8,1,10])
plt.plot(ypoints, marker = 'o')
plt.show()
```

```python
# Mark eacg point with a star
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3,8,1,10])
plt.plot(ypoints,marker = '*')

plt.show()
```

```python
# Format String
# Ex: Mark each point with a circle
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3,8,1,10])
plt.plot(ypoints, 'o:r')
plt.show()
```

```python
# Marker Size
# Ex: Set the size of the marker to 20
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3,8,1,10])
plt.plot(ypoints, marker = 'o', ms=20)
plt.show()
```

```python
# Marker Color
# Set the edge color to red
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3,8,1,10])
plt.plot(ypoints,marker='o', ms=20, mec='r')
plt.show()
```

```python
# Ex: Set the color of both the edge and the face to red
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3,8,1,10])
plt.plot(ypoints,marker='o',ms=20,mec='r',mfc='r')
plt.show()
```

```python
# You can also use Hexadecimal color values
# Ex: Mark each point a beautiful green color
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3,8,1,10])
plt.plot(ypoints, marker='o', ms=20,mec='#4CAF50',mfc='#4CAF50')
plt.show()
```

```python
# Or any of the 140 supported color names.
# Ex: Mark each point with the color named "hotpink".
import matplotlib.pyplot as plt
import numpy as np

ypoints=np.array([3,8,1,10])
plt.plot(ypoints, marker='o',ms=20,mec='hotpink',mfc='hotpink')
plt.show()
```
