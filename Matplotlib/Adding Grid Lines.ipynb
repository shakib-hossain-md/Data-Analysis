```python
# With pyplot,you can use the grid() function to add grid lines to the plot
# Ex: Add grid lines to athe plot
import matplotlib.pyplot as plt
import numpy as np
x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Colorie Burneg")
plt.plot(x,y)

plt.grid()
plt.show()
```

```python
# Specify which grid lines to display
# You can use the axis parameters in the grid() function to specify which grid line to display
# Ex: Display only grid line for the x-axis
import matplotlib.pyplot as plt
import numpy as np
x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])

plt.title("Sports Watch Data")
plt.xlabel("Average pulse")
plt.ylabel("Colorie Burn")
plt.plot(x,y)
plt.grid(axis='x')
plt.show()
```

```python
# Set line properties for the grid
# You can also set the line properties of the grid,like: grid(color=' ',linestyle=' ',linewidth=number)
# Ex: Set the line properties of the grid 
import matplotlib.pyplot as plt
import numpy as np
x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Colorie Burnage")
plt.plot(x,y)
plt.grid(color='green',linestyle='--',linewidth=0.5)
plt.show()
```
